Import the numpy module as np and the matplotlib.pyplot package as plt.

Create a numpy array
 containing the values for the natural logarithm function using the np.log(x) function. Create a numpy array
 containing the values for the exponential function using the np.exp(x) function.

 vectors to create two scatter plots of the points you've created.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of 50 points ranging from 0 to 1
x = np.linspace(0, 1, 50)

# Create numpy arrays for the natural logarithm and exponential functions
y_log = np.log(x + 1e-10)  # small value added to avoid log(0)
y_exp = np.exp(x)

# Create two scatter plots
plt.scatter(x, y_log, label='Natural Log')
plt.scatter(x, y_exp, label='Exponential')

# Display the plot
plt.show()
```

Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.

```python
x_grid = np.linspace(-6.5, 6.5, 100)

y_sin = np.sin(x_grid)
y_cos = np.cos(x_grid)

plt.plot(x_grid, y_sin, label='Sine')
plt.plot(x_grid, y_cos, label='Cosine')

plt.xlabel('x values')
plt.ylabel('y values')

plt.title('Sine and Cosine Functions')

plt.legend(loc='lower left')

plt.show()
```

Q4. Optimization is at the core of data science and statistics: Picking the best fit or estimate according to some desirable criteria (e.g. Maximum Likelihood Estimation). In this question, you're going to write a function that finds the highest possible value for some function, and returns the value of the function and the maximizing value.

Write a function that creates a grid from some value  𝑎  to a second value  𝑏  with  𝑁  steps, computes the value of a function  𝑓  on that grid, and then finds the maximizers of the function. Have it return a dictionary with the maximum value and the maximizers themselves.
Use your function to maximize  𝑓(𝑥)=100−2𝑥2+3𝑥  for values of  𝑎=−1 ,  𝑏=1 , and for values of  𝑁=3 ,  𝑁=10 ,  𝑁=100 ,  𝑁=1000  and  𝑁=5000 .
The true maximizer of this function is  .75  and the true maximum value is  101.125 . Is that what you got in the previous step? Why not? How does the quality of the maximization depend on  𝑁 ? How does your computed answer change as  𝑁  gets larger?

```python
import numpy as np
def f(x):
    return 100 - 2*x**2 + 3*x

def maximize_function(a, b, N):
    grid = np.linspace(a, b, N)
    values = f(grid)

    max_value = np.max(values)
    maximizer = grid[np.argmax(values)]
    return {"max_value": max_value, "maximizer": maximizer}

results = {}
for N in [3, 10, 100, 1000, 5000]:
    results[N] = maximize_function(-1, 1, N)

for N, result in results.items():
    print(f"N = {N}:")
    print(f"  Maximum value: {result['max_value']}")
    print(f"  Maximizer: {result['maximizer']}\n")
```
